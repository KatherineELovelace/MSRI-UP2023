```python
import math

radius = 1
area = math.pi*radius**2
```

```python
print(area)
```

```python
radii = [1 , 2 ,3 ,4]
```

```python
areas = []

for r in radii:
    area = math.pi*r**2
    areas.append(area)
```

```python
from matplotlib import pyplot as plt

%matplotlib inline

plt.plot(radii, areas, '-*',  markersize = 10 );
```

```python
def plot_disk_area(radii):
    areas = []
    for r in radii:
        area = math.pi*r**2
        areas.append(area)
    
    plt.plot(radii, areas, '-*',  markersize = 10 )
```

```python
plot_disk_area(radii)
```
